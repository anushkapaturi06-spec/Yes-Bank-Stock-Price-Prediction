The highest closing prices were recorded around 2017–2018. After this period, the yearly maximum closing price declined sharply.

This chart helps compare the lowest stock price reached in each year.

#### Chart - 20

````python
# Chart - 20 visualization code
# Rolling Standard Deviation (6-Month Risk Trend)

d['Rolling_STD'] = d['Close'].rolling(window=6).std()
plt.figure(figsize=(12,6))
plt.plot(d['Date'], d['Rolling_STD'], color='purple')
plt.title('6-Month Rolling Standard Deviation of Closing Price')
plt.xlabel('Year')
plt.ylabel('Rolling Standard Deviation')
plt.grid(True)
plt.show()
````

##### 1. Why did you pick the specific chart?

Rolling standard deviation measures how much the closing price varies over time, making it useful for analyzing changes in market risk.

##### 2. What is/are the insight(s) found from the chart?

The rolling standard deviation increases during periods of large stock price movements, indicating higher volatility and uncertainty in the market.
